```python
from datasets import load_dataset
import pandas as pd
```

```python
dataset = load_dataset("openai/MMMLU")
```

```python
df = dataset['test'].to_pandas()
```

```python
df
```

```python
df = df.iloc[70210:84252]
```

```python
df_law = df[df['Subject'] == 'professional_law']
df_law['Question']
df_law.to_csv('professional_law_questions.csv', index=False)
```

```python
df['Subject'].value_counts()
```

```python
df_law = df[(df['Subject'] == 'professional_law') | (df['Subject'] == 'jurisprudence')]
df_law.shape
```

```python
df_law['Reply'] = df_law.apply(lambda row: row[row['Answer']], axis=1)
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
df_law=df_law[['Question','Reply']]
```

```python
df_law.sample(5)
```

```python
df_law.to_csv('law.csv', index=False)
```

```python
df_medicine = df[(df['Subject'] == 'professional_medicine') | (df['Subject'] == 'clinical_knowledge') | (df['Subject'] == 'college_medicine') | (df['Subject'] == 'medical_genetics')]
df_medicine.shape
```

```python
df_medicine['Reply'] = df_medicine.apply(lambda row: row[row['Answer']], axis=1)
```

```python
df_medicine=df_medicine[['Question','Reply']]
```

```python
df_medicine.sample(5)
```
